```python
# All the imports for the task
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Loading the test image
image_path = '../Lab_1/img.jpg'
img = cv2.imread(image_path)

# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plot the image
plt.figure(figsize=(10, 8))
plt.imshow(img)
plt.title('Original test image in RGB')
plt.axis('off')
plt.show()
```

```python
# Split the image into the three colour channels
red, green, blue = cv2.split(img)

# Compose the image in the RGB colour space
img_rgb = cv2.merge([red, green, blue])

# Compose the image in the RBG colour space
img_rbg = cv2.merge([red, blue, green])

# Compose the image in the GRB colour space
img_grb = cv2.merge([green, red, blue])

# Compose the image in the BGR colour space
img_bgr = cv2.merge([blue, green, red])
```

```python
# Create the collage
out1 = np.hstack([img_rgb, img_rbg])
out2 = np.hstack([img_grb, img_bgr])
out = np.vstack([out1, out2])

# Plot the collage
plt.imshow(out)
plt.axis(False)
```

```python
# Create a figure and axis for plotting
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot each image
axs[0, 0].imshow(img_rgb)
axs[0, 0].set_title('RGB image')
axs[0, 0].axis('off')

axs[0, 1].imshow(img_rbg)
axs[0, 1].set_title('RBG image')
axs[0, 1].axis('off')

axs[1, 0].imshow(img_grb)
axs[1, 0].set_title('GRB image')
axs[1, 0].axis('off')

axs[1, 1].imshow(img_bgr)
axs[1, 1].set_title('BGR image')
axs[1, 1].axis('off')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
