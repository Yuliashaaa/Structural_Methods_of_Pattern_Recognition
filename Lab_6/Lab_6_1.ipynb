## Face Detection via dlib

Dlib is a general purpose cross-platform software library that contains many useful tools. In particular, it includes a trained DNN for face detection.

```python
# All the imports for the task
import cv2
import dlib
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Loading the test image
image_path = '../Lab_6/test_face.jpg'
image = cv2.imread(image_path)

# Convert BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Plot the image
plt.figure(figsize=(12, 8))
plt.imshow(image)
plt.title('Original Test Image')
plt.axis('off')
plt.show()
```

```python
def dlib_face_detector(img: np.ndarray):
    """
    Detects faces in the input image using dlib's frontal face detector.

    Args:
        img (np.ndarray): Input image.

    Returns:
        Displays the input image with bounding boxes around the detected faces.
    """
    # Load the dlib face predictor
    face_detector = dlib.get_frontal_face_detector()
    
    # Run the predictor on your image
    detected_faces = face_detector(img)
    
    # Draw bounding boxes around the detected faces and plot the image
    plt.figure(figsize=(12, 8))
    for i, face in enumerate(detected_faces):
        x, y, w, h = face.left(), face.top(), face.width(), face.height()
        color = tuple(int(c) for c in np.random.randint(0, 256, size=3))  # Random color for each face
        cv2.rectangle(img, pt1=(x, y), pt2=(x + w, y + h), color=color, thickness=4)
        plt.text(x, y, 'Face {}'.format(i + 1), color='green', fontsize=12)

    plt.imshow(img)
    plt.title('Detected Face(s)')
    plt.axis('off')
    plt.show()
```

```python
dlib_face_detector(image)
```
